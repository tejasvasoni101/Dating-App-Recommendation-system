NUMBERING AND ADDING A ID COLUMN

```python
list_of = []
# Iterate over a sequence of numbers from START to n
start = 1001
for i in range(start,6001):  #11000
    # In each iteration, add ELEMENT to the main list
    list_of.append(str(i))  #DATA TYPE IS STRING

df2 = pd.DataFrame(columns = list_of, 
                   index = list_of)   
#df2.drop(["userID"], inplace = True)
```

```python
Reco_data_output = df2    #RECO IS EMPTY NOW
Reco_data_output.to_csv('Reco_data_output.csv',index=True)
Reco_data_output.head()
```
